```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv
import time
from datetime import datetime,timezone
from pytz import timezone


def read_csv(fileName):
    data = []
    f = open(fileName,'r')
    spamreader = csv.DictReader(f)
    for row in spamreader:
        data.append(row)
    return data

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
#         print('Message published successfully. Data: ' + str(data))
#         print("-----------------------------------------------------------")
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
def divideData(data):
    data_4 = []
    data_5 = []
    data_6 = []
    data_7 = []
    data_8 = []
    for i in data:
        if i['machine'] == '4':
            data_4.append(i)
        elif i['machine'] == '5':
            data_5.append(i)
        elif i['machine'] == '6':
            data_6.append(i)
        elif i['machine'] == '7':
            data_7.append(i)
        elif i['machine'] == '8':
            data_8.append(i)        
    return data_4,data_5,data_6,data_7,data_8

if __name__ == '__main__':
   
    topic = 'Streaming_Linux_memory'
    cRows = read_csv('Streaming_Linux_memory.csv') 
    print('Publishing records..')
    producer = connect_kafka_producer() 
    machine_4,machine_5,machine_6,machine_7,machine_8 = divideData(cRows)
    start_num_4,start_num_5,start_num_6,start_num_7,start_num_8 = 0,0,0,0,0
    start_num_4_y,start_num_5_y,start_num_6_y,start_num_7_y,start_num_8_y = 0,0,0,0,0
    data_y=[]
    while True:
        random_4_y,random_5_y,random_6_y,random_7_y,random_8_y = random.randint(0,5),random.randint(0,5)\
            ,random.randint(0,5),random.randint(0,5),random.randint(0,5)
        random_4,random_5,random_6,random_7,random_8 = random.randint(20,80),random.randint(20,80)\
            ,random.randint(20,80),random.randint(20,80),random.randint(20,80)
        data = []
        ts = int(datetime.now(timezone('UTC')).timestamp())
        if (start_num_4 + random_4) > len(machine_4):
            start_num_4 = 0
        if (start_num_5 + random_5) > len(machine_5):
            start_num_5 = 0
        if (start_num_6 + random_6) > len(machine_6):
            start_num_6 = 0
        if (start_num_7 + random_7) > len(machine_7):
            start_num_7 = 0
        if (start_num_8 + random_8) > len(machine_8):
            start_num_8 = 0
         
        for a in range(start_num_4,start_num_4+random_4):
            machine_4[a]['ts'] = ts
            data.append(machine_4[a])
        start_num_4 += random_4
        for b in range(start_num_5,start_num_5+random_5):
            machine_5[b]['ts'] = ts
            data.append( machine_5[b])
        start_num_5 += random_5
        for c in range(start_num_6,start_num_6+random_6):
            machine_6[c]['ts'] = ts
            data.append(machine_6[c])
        start_num_6 += random_6
        for d in range(start_num_7,start_num_7+random_7):
            machine_7[d]['ts'] = ts
            data.append(machine_7[d])
        start_num_7 += random_7
        for e in range(start_num_8,start_num_8+random_8):
            machine_8[e]['ts'] = ts
            data.append(machine_8[e])
        start_num_8 += random_8
        data.extend(data_y)
        publish_message(producer,topic,data)
        sleep(10)
        ts = int(datetime.now(timezone('UTC')).timestamp())
        data_y.clear
        if (start_num_4_y + random_4_y) > len(machine_4):
            start_num_4_y = 0
        if (start_num_5_y + random_5_y) > len(machine_5):
            start_num_5_y = 0
        if (start_num_6_y + random_6_y) > len(machine_6):
            start_num_6_y = 0
        if (start_num_7_y + random_7_y) > len(machine_7):
            start_num_7_y = 0
        if (start_num_8_y + random_8_y) > len(machine_8):
            start_num_8_y = 0
         
        for a_y in range(start_num_4_y,start_num_4_y+random_4_y):
            machine_4[a_y]['ts'] = ts
            data_y.append(machine_4[a_y])
        start_num_4_y += random_4_y
        
        for b_y in range(start_num_5_y,start_num_5_y+random_5_y):
            machine_5[b_y]['ts'] = ts
            data_y.append( machine_5[b_y])
        start_num_5_y += random_5_y

        for c_y in range(start_num_6_y,start_num_6_y+random_6_y):
            machine_6[c_y]['ts'] = ts
            data_y.append(machine_6[c_y])
        start_num_6_y += random_6_y

        for d_y in range(start_num_7_y,start_num_7_y+random_7_y):
            machine_7[d_y]['ts'] = ts
            data_y.append(machine_7[d_y])
        start_num_7_y += random_7_y

        for e_y in range(start_num_8_y,start_num_8_y+random_8_y):
            machine_8[e_y]['ts'] = ts
            data_y.append(machine_8[e_y])
        start_num_8_y += random_8_y

        
            
```
